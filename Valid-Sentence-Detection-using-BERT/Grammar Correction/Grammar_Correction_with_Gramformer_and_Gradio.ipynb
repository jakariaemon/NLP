```python
from gramformer import Gramformer
```

```python
!python -m spacy download en_core_web_sm
```

```python
gf = Gramformer(models=1, use_gpu=False) #1=corrector, 2=detector
```

```python
gf.correct('I taken dinner') 
```

```python
sentences = [
                'My name a Lilian',
                'I love watching an action movie',
                'I has a laptop',
                'My phone battery a spoilt'
]
```

```python
for sentence in sentences:
    res = gf.correct(sentence)
    print(res)
```

**Deploying the model as an application using Gradio**

```python
!pip install gradio
```

```python
import gradio as gr
```

```python
def correct(sentence):
    res = gf.correct(sentence) 
    return res
```

```python
app_inputs = gr.inputs.Textbox(lines=3, placeholder="Enter text to know its toxicity...")
```

```python
interface = gr.Interface(fn=correct, 
                        inputs=app_inputs,
                         outputs='text', 
                        title='Hi there, I\'m Gramformer')
```

```python
interface.launch()
```
